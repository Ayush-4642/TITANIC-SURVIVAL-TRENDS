```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")  # âœ… Apply .head(40) after loading
df.head(50)
```

```python
order_1=["Cherbourg","Southampton","Queenstown"]
sns.barplot(y="age", x="embark_town", hue="sex", data=df,order=order_1,errorbar=("ci",50))
plt.title("Embarkation Town vs. Age: Gender-Based Comparison")
plt.show()
```

```python
sns.countplot(x="class",hue="sex",data=df)
plt.title("Passenger Class Distribution by Gender")
plt.show()
```

```python
sns.countplot(x="survived",hue="sex",data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.barplot(x="survived", y="age",hue="sex",  data=df, errorbar=None)
plt.title("Average Age by Survival Status and Gender")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.histplot(data=df, x="fare", hue="sex",multiple="dodge", bins=[0,100,200,300,400,500])
plt.show()
```

```python
sns.histplot(data=df, x="fare", hue="sex", multiple="dodge", stat="density", kde=True)
plt.title("Fare Distribution by Gender with Density Curves")
plt.show()
```

```python
sns.barplot(x="age", y="sex", hue="who", data=df)
plt.title("Average Age by Gender and Passenger Type")
plt.show()
```

```python
sns.barplot(x="age", y="alive", hue="who", data=df)
plt.title("Who Lived and Died: Age Analysis by Group")
plt.show()
```

```python
sns.displot(x="alone", hue="sex", data=df, multiple="dodge")
plt.title("Alone Status by Gender")
plt.show()
```

```python
m={"male":"o","female":"s"}
sns.scatterplot(   x="fare", y="age", data=df,hue="sex", palette="gist_ncar",style="sex", size="sex", sizes=(50, 60),alpha=1, markers=m)
plt.title("Fare vs Age by Gender with Custom Markers")
plt.show()
       
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Missing Data Heatmap After Imputation")
plt.show()
```

```python
df = df.drop(columns="deck")
```

```python
df.isnull().sum()
```

```python
df[["age", "fare", "embark_town"]].head(50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=True, cmap="YlGnBu")
plt.title("Missing Data Heatmap After Imputation")
plt.show()
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df["embark_town"] = df["embark_town"].fillna(df["embark_town"].mode()[0])
```

```python
sns.heatmap(df.isnull(), cbar=True, cmap="YlGnBu")
plt.title("Missing Data Heatmap After Final Imputation")
plt.show()
```

Outline your steps:
- Handled missing age with median (28.0)
- Filled fare using mean
- Filled embark_town with mode (Southampton)
- Dropped deck due to heavy missingness
- Verified cleanliness with heatmaps
â€œBefore exploring patterns, it was critical to make sure every row had a story to tell.â€
